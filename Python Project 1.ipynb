```python
#Importing pandas and numpy packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Creating dataframe
dfschools = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#Exploring 5 first rows of data
dfschools.head()
```

```python
#Finding data type for each column of the dataframe 
dfschools.info()
```

```python
#Finding na values
dfschools.isna().sum()
```

```python
#Finding unique values in Regents Exam column 
dfschools['Regents Exam'].unique()
```

```python
#Removing na values 
dfschools.dropna()
```

```python
#Finding unique values for each column
dfschools.nunique()
```

```python
#Removing all duplicate values 
dfschools.drop_duplicates()
```

```python
#Removing all rows without a mean score
dfschools = dfschools.drop(dfschools[dfschools['Mean Score'] == 's'].index)
dfschools
```

```python
#Adding new column to identify the code of the borough 
dfschools['Code'] = dfschools['School DBN'].str[2]
dfschools.head()
```

```python
#Categorizing codes for boroughs
def categorize_borough(code):
    if code == 'M':
        return 'Manhattan'
    elif code == 'K':
        return 'Brooklyn'
    elif code == 'X':
        return 'Bronx'
    elif code == 'Q':
        return 'Queens'
    else:
        return 'Staten Island'
    
#Creating new columns for boroughs 
dfschools['Borough'] = dfschools['Code'].apply(categorize_borough)

dfschools.head()
```

```python
dfschools.info()
```

```python
#Changing data type for columns to numeric 
dfschools['Mean Score'] = pd.to_numeric(dfschools['Mean Score'])
dfschools['Number Scoring Below 65'] = pd.to_numeric(dfschools['Number Scoring Below 65'])
dfschools['Percent Scoring Below 65'] = pd.to_numeric(dfschools['Percent Scoring Below 65'])
dfschools['Number Scoring 65 or Above'] = pd.to_numeric(dfschools['Number Scoring 65 or Above'])
dfschools['Percent Scoring 65 or Above'] = pd.to_numeric(dfschools['Percent Scoring 65 or Above'])
dfschools['Number Scoring 80 or Above'] = pd.to_numeric(dfschools['Number Scoring 80 or Above'])
dfschools['Percent Scoring 80 or Above'] = pd.to_numeric(dfschools['Percent Scoring 80 or Above'])
```

```python
#Changing data type for columns to string
dfschools['School Name'] = dfschools['School Name'].astype("string")
dfschools['School Level'] = dfschools['School Level'].astype("string")
dfschools['Regents Exam'] = dfschools['Regents Exam'].astype("string")
dfschools['Code'] = dfschools['Code'].astype("string")
dfschools['Borough'] = dfschools['Borough'].astype("string")
```

```python
dfschools.info()
```

```python
#Mean score per borough

boroughs_meanscore = dfschools.groupby(['Borough'])['Mean Score'].mean()
```

```python
dfbar = pd.DataFrame(data=boroughs_meanscore, columns = ['Mean Score'])
dfbar
```

```python
#Mean score per Borough bar plot

dfbar.plot.bar()
plt.title('Mean Score per Borough', size=15)
plt.xlabel('Borough', size = 12)
plt.ylabel('Mean Score', size = 12)
plt.xticks(rotation=0);
```

```python
#Counting number scoring below 65 per borough
Borough_below65 = dfschools.groupby(['Borough'])['Number Scoring Below 65'].sum()
Borough_below65
```

```python
#Counting number scoring 65 or above per borough
Borough_above65 = dfschools.groupby(['Borough'])['Number Scoring 65 or Above'].sum()
Borough_above65
```

```python
#Counting number scoring 80 or above per borough
Borough_above80 =  dfschools.groupby(['Borough'])['Number Scoring 80 or Above'].sum()
Borough_above80
```

```python
#Dataframes for scoring below 65
dfbelow65 = pd.DataFrame(data = Borough_below65, columns =['Number Scoring Below 65'])
dfbelow65
```

```python
#Dataframes for scoring above 65
dfabove65 = pd.DataFrame(data = Borough_above65, columns =['Number Scoring 65 or Above'])
dfabove65
```

```python
#Dataframes for scoring above 80
dfabove80 = pd.DataFrame(data = Borough_above80, columns =['Number Scoring 80 or Above'])
dfabove80
```

```python
#Joining scoring dataframes
frames = [dfbelow65, dfabove65, dfabove80]
dfscorings1 = pd.merge(dfbelow65,dfabove65, on='Borough', how='inner')
```

```python
#Joining scoring dataframes
frames = [dfscorings1, dfabove80]
dfscorings2 = pd.merge(dfscorings1, dfabove80, on='Borough', how='inner')
dfscorings2
```

```python
#Number of students evaluated per Boroughs heatmap
dfscorings2.style.background_gradient(cmap='Greens');
```
